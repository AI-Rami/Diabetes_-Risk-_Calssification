# **Outlier Summary by Percentile**

---
This is to find out how many outliers do we have in the columns BMI, MentHlth, and PhysHlth using percentile 99th and 1 th


```python
outlier_features = ['BMI', 'MentHlth', 'PhysHlth']

for col in outlier_features:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    num_lower = (df[col] < lower).sum()
    num_upper = (df[col] > upper).sum()
    num_outliers = num_lower + num_upper
    percent_outliers = 100 * num_outliers / len(df)
    print(f"{col}:")
    print(f"  1st percentile value: {lower:.2f}")
    print(f"  99th percentile value: {upper:.2f}")
    print(f"  Outliers below 1st percentile: {num_lower}")
    print(f"  Outliers above 99th percentile: {num_upper}")
    print(f"  Total outliers: {num_outliers} ({percent_outliers:.2f}%)\n")
```

# **Outlier Summary by IQR Rule**

---
This is to find out how many outliers do we have in the columns BMI, MentHlth, and PhysHlth using IQR rule.



```python
for col in ['MentHlth', 'PhysHlth','BMI']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col}: Outliers by IQR method = {outliers}")
```

# **BMI Outlier Detection and Summary**

```python
# this is to calculate Q1, Q3, and IQR for BMI
Q1 = df['BMI'].quantile(0.25)
Q3 = df['BMI'].quantile(0.75)
IQR = Q3 - Q1

# here we want to define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# this is to find outliers
outliers = df[(df['BMI'] < lower_bound) | (df['BMI'] > upper_bound)]

# calculating of number and percentage of outliers
num_outliers = outliers.shape[0]
percent_outliers = 100 * num_outliers / len(df)

print(f"BMI outlier calculation using IQR:")
print(f"  Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
print(f"  Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")
print(f"  Outliers found: {num_outliers} ({percent_outliers:.2f}%)")
```
